```python
qus = df.columns
qus=qus.drop(["Time",'Time Code','Country Name', 'Country Code'])
```

```python
df.dropna(subset=['Time', 'Country Name'],inplace=True)
for i in qus:
    df[i].replace('..',np.nan,inplace=True)
    
qus= list(qus)
for i in qus:
    if df[i].isnull().sum() > 100:
        qus.remove(i)
df[qus].isnull().sum()
```

```python
for i in qus:
    df[i].fillna(method='ffill',inplace=True)
    df[i].fillna(method='bfill',inplace=True)
    df[i].astype('float64')
```
